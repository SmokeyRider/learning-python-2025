```python
import pandas as pd

df = pd.read_csv("https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user", sep="|")
df.head()

print(df.shape)
print(df.columns)
print(df.describe())
print(df)
```

```python
df.occupation.value_counts()
df['age'].hist(bins=20)
```

```python
import matplotlib.pyplot as plt

# Plot histogram of the 'Age' column
df['age'].hist(bins=20, color="blue", alpha=0.5)
plt.title('Age Histogram')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()

# Plot bar chart of the 'Occupation' column
df['occupation'].value_counts().plot(kind='bar', color='blue', alpha=0.5)
plt.title('Occupation Bar Chart')
plt.xlabel('Occupation')
plt.ylabel('Count') 
plt.show()
```

```python
# Manually fixing line endings in Jupyter Notebook
def convert_to_unix_line_endings(file_path):
    with open(file_path, 'rb') as file:
        content = file.read()
    content = content.replace(b'\n', b'\r\n')
    with open(file_path, 'wb') as file:
        file.write(content)
convert_to_unix_line_endings('Jupyter_nb.ipynb')
```
