## Get to know the basics of matplotlib

```python
import matplotlib.pyplot as plt
import pandas as pd

df = pd.read_csv("https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user", sep="|")

#df.occupation.value_counts()
df['age'].hist(bins=50)
plt.title('Age Histogram: 50 bins')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()

# Plot histogram of the 'Age' column
df['age'].hist(bins=20, color="blue", alpha=0.5)
plt.title('Age Histogram: 20 bins')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

### More examples: bar charts

```python
import matplotlib.pyplot as plt
import pandas as pd

df = pd.read_csv("https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user", sep="|")

# Plot bar chart of the 'Occupation' column
df['occupation'].value_counts().sort_values(ascending=True).plot(kind='barh', color='blue', alpha=0.5)
plt.title('Horizontal Bar Chart for Occupation')
plt.xlabel('Count') 
plt.ylabel('Occupation')
plt.show()
```

### A little hack in case git barks at you about the wrong EOL characters

```python
# Manually fixing line endings in Jupyter Notebook
def convert_to_windows_line_endings(file_path):
    with open(file_path, 'rb') as file:
        content = file.read()
    content = content.replace(b'\n', b'\r\n')
    with open(file_path, 'wb') as file:
        file.write(content)
convert_to_windows_line_endings('Jupyter_nb.ipynb')
```
