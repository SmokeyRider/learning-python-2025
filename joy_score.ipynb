# Create Activity Dataset
Create a pandas DataFrame containing activities with their names, happiness scores (1-10), and hip-health scores (1-10).

```python
import pandas as pd

activity_df = pd.DataFrame([
    ["Cenote walk", 5, 3],
    ["Pool Swim", 4, 6],
    ["Beach walk", 6, 1], 
    ["Pool Lounging", 3, 5],
    ["Bed lounging", 1, 9],
    ["River Swim", 8, 4]
], columns=["Activity Name", "Happiness Score", "Hip-Health Score"])

# Display the DataFrame
activity_df
```

# Import and Setup Libraries
Import required libraries including pandas for data management and matplotlib/seaborn for visualization.

```python
# Import required libraries
import matplotlib.pyplot as plt
import seaborn as sns

# Set a style for the visualizations
sns.set(style="whitegrid")
```

# Visualize Activity Data
Create a scatter plot using happiness scores on x-axis and hip-health scores on y-axis, with activity names as data points.

```python
# Create a scatter plot for visualizing the activity data
plt.figure(figsize=(10, 6))

# Scatter plot with Happiness Score on x-axis and Hip-Health Score on y-axis
scatter = sns.scatterplot(
    x="Happiness Score",
    y="Hip-Health Score",
    data=activity_df,
    s=100,  # Size of the markers
    color="blue"
)

# Annotate each point with the activity name
for i in range(activity_df.shape[0]):
    plt.text(
        x=activity_df["Happiness Score"][i] + 0.1,  # Slight offset for better visibility
        y=activity_df["Hip-Health Score"][i],
        s=activity_df["Activity Name"][i],
        fontsize=10
    )

# Add labels and title
plt.title("Activity Comparison: Happiness vs Hip-Health", fontsize=14)
plt.xlabel("Happiness Score", fontsize=12)
plt.ylabel("Hip-Health Score", fontsize=12)

# Show the plot
plt.show()
```

# Add Labels and Styling
Add title, axis labels, and customize the plot appearance. Add activity names as annotations for each point.

```python
# Add title, axis labels, and customize the plot appearance. Add activity names as annotations for each point.

# Update the scatter plot with additional styling and labels
plt.figure(figsize=(10, 6))

# Scatter plot with Happiness Score on x-axis and Hip-Health Score on y-axis
scatter = sns.scatterplot(
    x="Happiness Score",
    y="Hip-Health Score",
    data=activity_df,
    s=150,  # Increase marker size for better visibility
    color="green",
    edgecolor="black"  # Add edge color to markers
)

# Annotate each point with the activity name
for i in range(activity_df.shape[0]):
    plt.text(
        x=activity_df["Happiness Score"][i] + 0.15,  # Adjust offset for better visibility
        y=activity_df["Hip-Health Score"][i] + 0.1,
        s=activity_df["Activity Name"][i],
        fontsize=11,
        fontweight="bold",
        color="darkred"
    )

# Add labels and title with styling
plt.title("Activity Comparison: Happiness vs Hip-Health", fontsize=16, fontweight="bold", color="darkblue")
plt.xlabel("Happiness Score", fontsize=14, fontweight="bold", color="darkgreen")
plt.ylabel("Hip-Health Score", fontsize=14, fontweight="bold", color="darkgreen")

# Customize grid and background
plt.grid(color="gray", linestyle="--", linewidth=0.5)
plt.gca().set_facecolor("whitesmoke")

# Show the updated plot
plt.show()
```
